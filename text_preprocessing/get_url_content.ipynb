## Write a function that takes as input an url and return page title, text, publish date, and authors 

```python
from newspaper import Article  # pip3 install newspaper3k

def get_url_content(url):
    """
    """
    art=Article(url)
    art.download()
    art.parse()
    content={
        "url":url,
        "title":art.title,
        "text":art.text,
        "authors":art.authors,
        "publish_date":art.publish_date}
    return content
```

```python
url="https://www.defuwaxu.com/2019/08/30/waxtaanu-saada-kan-ak-bubakar-boris-joob/"

content=get_url_content(url)

from pprint import pprint

pprint(content)
    
```
