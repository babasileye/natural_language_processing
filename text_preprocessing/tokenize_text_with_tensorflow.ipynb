```python
text_file="data/les_miserables_victor_hugo.txt"

with open(text_file,'r') as f:
    lines = f.readlines()
    
with open(text_file,'r') as f:
    text = f.read()
```

## Create tokenizer

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer=Tokenizer(num_words=20000,
                    filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
                    lower=True,
                    split=' ',
                    char_level=False,
                    oov_token='<unk>')
```

```python
tokenizer.fit_on_texts(lines)
```

```python
tokenizer.__dict__.keys()
```

```python
tokenizer.word_index
```

```python
tokenizer.index_word
```

```python
# Encode training data sentences into sequences
sequences = tokenizer.texts_to_sequences(lines)
print(lines[0],sequences[0])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_sequences = pad_sequences(sequences, padding='post', truncating='pre', maxlen=15)
padded_sequences[0]
```

```python
reconstructed_texts=tokenizer.sequences_to_texts(padded_sequences)
print(padded_sequences[0],reconstructed_texts[0])
```
