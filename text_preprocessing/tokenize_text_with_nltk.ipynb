```python
text_file="data/les_miserables_victor_hugo.txt"

with open(text_file,'r') as f:
    text = f.read()

number_of_characters=len(text)
print(f'number of characters is {number_of_characters}')
print(f'first ten characters: {text[0:9]}')
```

```python
with open(text_file) as f:
    lines = f.readlines()

number_of_lines=len(lines)
print(f'number of lines is {number_of_lines}')
print(f'first line: {lines[0]}')
```

```python
# pour installer nltk: pip3 install nltk
from nltk.tokenize import word_tokenize,sent_tokenize
#word_tokenize: pour segmenter un  texte en mots
#sent_tokenize: pour segmenter un  texte en phrases

sentences=sent_tokenize(text)
words=word_tokenize(text)

print(f'First sentence: {sentences[0]}')
```

```python
print(f'words: {words[3]}')
```

```python
print(f'First word: {words[0]}')
```

## Store sentences into a pandas dataframe 

```python
import pandas as pd

data=[]
for sentence in sentences:
    data.append({"sentence":sentence,"book":text_file})
df=pd.DataFrame.from_dict(data)
print(f'number of sentences is {df.shape[0]}')
df.head()
    
```

## Counts word occurences, and display occurences as a word cloud

```python
from collections import defaultdict
word_counts = defaultdict(int)

for word in words:
    word_counts[word.lower()] +=1
```

### Use the python package WordCloud to display word cloud

```python
from wordcloud import WordCloud #pip3 install wordcloud
import matplotlib.pyplot as plt

wc = WordCloud(background_color="black")
wc.generate_from_frequencies(word_counts)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
